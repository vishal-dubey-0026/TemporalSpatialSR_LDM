```python
import glob
import os
os.chdir("/content/drive/MyDrive/")
dataset_folder = "second-annual-data-science-bowl-v2"
tag = "3d"
num_adj_slice = 1
num_slices = 2*num_adj_slice + 1
temp_frame_count = 30
root_path = "/content"
path = f"Diffusion_SR/SinSR/traindata/{dataset_folder}-{tag}"
if not os.path.exists(path):
  os.makedirs(path)

path = f"Diffusion_SR/SinSR/testdata/{dataset_folder}-{tag}"
if not os.path.exists(path):
  os.makedirs(path)


splits = ['test', 'train'] # 'validate'
for split in splits:
  split_folder = f"{dataset_folder}/{split}/{split}"
  if not os.path.exists(split_folder): continue
  patients_name = os.listdir(split_folder)
  for pat_name in patients_name:
    print(f"{pat_name}")
    pat_folder = f"{split_folder}/{pat_name}"
    #######################
    slices_name = sorted([name for name in os.listdir(f"{pat_folder}/study") if "sax_" in name], key=lambda x: int(x.split("_")[-1]))
    if len(slices_name) < num_slices: continue
    for slice_idx, slice_name in enumerate(slices_name):
      if (slice_idx == 0) or (slice_idx == len(slices_name) - 1): continue
      if "sax_" not in slice_name: continue
      slice_time_li = []
      for adj_t in range(-num_adj_slice, num_adj_slice + 1):

          slice_folder = f"{pat_folder}/study/{slices_name[slice_idx + adj_t]}"
          slice_time_li.append(sorted(glob.glob(f"{slice_folder}/*.dcm")))
          if len(slice_time_li[-1]) != temp_frame_count:
            print(f"{slice_folder} has {len(slice_time_li[-1])} timesteps")
            slice_time_li.pop()
            break
      if len(slice_time_li) != num_slices: continue
      zipped_slice_time_li = []
      for time_idx in range(len(slice_time_li[0])):

          slice_time_li[0][time_idx] = slice_time_li[0][time_idx].replace(f"{root_path}/",f"")
          slice_time_li[1][time_idx] = slice_time_li[1][time_idx].replace(f"{root_path}/",f"")
          slice_time_li[2][time_idx] = slice_time_li[2][time_idx].replace(f"{root_path}/",f"")
          zipped_slice_time_li.append(f"{slice_time_li[0][time_idx]}\t{slice_time_li[1][time_idx]}\t{slice_time_li[2][time_idx]}")

      store_folder = 'traindata'
      #if split == 'test' and int(pat_name) > 800:
        #store_folder = 'testdata'

      with open(f"Diffusion_SR/SinSR/{store_folder}/{dataset_folder}-{tag}/{pat_name}_{slice_name}.txt", 'w') as file:
        for item in zipped_slice_time_li:
          file.write(item + '\n')
    #########################
```
