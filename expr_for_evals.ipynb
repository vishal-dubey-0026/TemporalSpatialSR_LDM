```python
import glob
import os
os.chdir("/content/drive/MyDrive/")
dataset_folder = "second-annual-data-science-bowl-v2"

path = f"Diffusion_SR/SinSR/traindata/{dataset_folder}"
if not os.path.exists(path):
  os.makedirs(path)

path = f"Diffusion_SR/SinSR/testdata/{dataset_folder}"
if not os.path.exists(path):
  os.makedirs(path)


splits = ['test', 'train'] # 'validate'
for split in splits:
  split_folder = f"{dataset_folder}/{split}/{split}"
  if not os.path.exists(split_folder): continue
  patients_name = os.listdir(split_folder)
  for pat_name in patients_name:
    print(f"{pat_name}")
    pat_folder = f"{split_folder}/{pat_name}"
    slices_name = os.listdir(f"{pat_folder}/study")
    for slice_name in slices_name:
      if "sax_" not in slice_name: continue
      slice_folder = f"{pat_folder}/study/{slice_name}"
      slice_time_li = sorted(glob.glob(f"{slice_folder}/*.dcm"))
      if len(slice_time_li) <= 10: continue
      store_folder = 'traindata'
      #if split == 'test' and int(pat_name) > 800:
        #store_folder = 'testdata'

      with open(f"Diffusion_SR/SinSR/{store_folder}/{dataset_folder}/{pat_name}_{slice_name}.txt", 'w') as file:
        for item in slice_time_li:
          file.write(item + '\n')
```
